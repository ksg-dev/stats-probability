# Percentiles

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()
```

```python
# 50th percentile - same as median
np.percentile(vals, 50)
```

```python
# 90th percentile - where 90% of the data is less than this given value
np.percentile(vals, 90)
```

```python
np.percentile(vals, 20)
```
